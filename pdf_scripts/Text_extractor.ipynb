```python
import PyPDF2
from PyPDF2 import PdfFileReader, PdfFileWriter
```

```python
file_path = 'C:/Users/LUGAPEDE/Documents/data_projects/IT/2021-10-05_Roadmap_all_countries - Copy.pdf'
```

```python
pdf = PdfFileReader(file_path)
with open ('all_countries.txt', 'w') as f:
    for page_num in range(pdf.numPages):
        #print('Page : {0}'.format(page_num))
        pageObj = pdf.getPage(page_num)
        
        try:
            txt = pageObj.extractText()
            print(''.center(100, '*'))
        except:
            pass
        else:
            f.write('Page {0}\n'.format(page_num +1))
            f.write(''.center(100,'*'))
            f.write(txt)
    f.close()
```

```python
# remove white spaces in the text
with open ('all_countries.txt', 'r') as f:
    lines = f.readlines()
lines = [line.replace('   ', '') for line in lines]
lines = [line.replace('  ', '') for line in lines]
lines = [line.replace('    ', '') for line in lines]

with open('all_countries.txt', 'w') as f:
    f.writelines(lines)
```
